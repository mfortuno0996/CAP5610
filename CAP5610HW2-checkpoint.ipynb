# Obtaining Titanic Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

train_df = pd.read_csv('CAP5610HW1/train.csv')
test_df = pd.read_csv('CAP5610HW1/test.csv')
print(train_df)
```

## Data Preprocessing
### Age


```python
data = [train_df, test_df] #combining training and test set to perform data preprocessing.

#Fill in all null values for age using random numbers between mean and std for training and test sets
for dataset in data:
    mean = train_df["Age"].mean()
    std = train_df["Age"].std() #using training set data for mean and std as training has more values
    is_null = dataset["Age"].isnull().sum() #outputs the size of total null values in Age feature
    rand_age = np.random.randint(mean - std, mean + std, size = is_null)
    age_fill = dataset["Age"].copy()
    age_fill[np.isnan(age_fill)] = rand_age # fills null values in Age feature with the random values created
    dataset["Age"] = age_fill
    dataset["Age"] = train_df["Age"].astype('int64') #converts age feature from float to int
train_df["Age"].isnull().sum() #verifies all null values are filled for training set

train_df
```

```python
data = [train_df, test_df]
for dataset in data: #converting discrete age to integer age groups 
    dataset['Age'] = dataset['Age'].astype(int)
    dataset.loc[ dataset['Age'] <= 10, 'Age'] = 0
    dataset.loc[(dataset['Age'] > 10) & (dataset['Age'] <= 18), 'Age'] = 1
    dataset.loc[(dataset['Age'] > 18) & (dataset['Age'] <= 24), 'Age'] = 2
    dataset.loc[(dataset['Age'] > 24) & (dataset['Age'] <= 30), 'Age'] = 3
    dataset.loc[(dataset['Age'] > 30) & (dataset['Age'] <= 36), 'Age'] = 4
    dataset.loc[(dataset['Age'] > 36) & (dataset['Age'] <= 42), 'Age'] = 5
    dataset.loc[(dataset['Age'] > 42) & (dataset['Age'] <= 66), 'Age'] = 6
    dataset.loc[ dataset['Age'] > 66, 'Age'] = 6
data
```

### Embarked

```python
for dataset in data:
    dataset["Embarked"].fillna("S", inplace = True)  #filling all NA values with the most common value "S"
    dataset.replace({'S':0, 'C':1, 'Q':2}, inplace = True) #converting string to numeric for Embarked Feature
train_df["Embarked"].isnull().sum() #verifies values are filled
train_df
```

### Sex

```python
for dataset in data:
    dataset.rename(columns = {'Sex' : 'Gender'}, inplace=True) #Replacing Sex with Gender
    dataset.replace({'male':0, 'female':1}, inplace=True) #replaces all instances of male and female with 0 and 1 respectively
data
```

### Fare

```python
for dataset in data:
    dataset['Fare'] = dataset['Fare'].fillna(0)
    dataset.loc[ dataset['Fare'] <= 7.91, 'Fare'] = 0
    dataset.loc[(dataset['Fare'] > 7.91) & (dataset['Fare'] <= 14.454), 'Fare'] = 1
    dataset.loc[(dataset['Fare'] > 14.454) & (dataset['Fare'] <= 31), 'Fare']   = 2
    dataset.loc[(dataset['Fare'] > 31) & (dataset['Fare'] <= 512.329), 'Fare']   = 3
    dataset['Fare'] = dataset['Fare'].astype(int)
data
```

## Removing Unnecessary Features

```python
#Removing PassengerID, Name, Cabin, and Ticket from the training set, as those do not have significant correlation with survival
train_df = train_df.drop(['PassengerId','Name','Cabin','Ticket'], axis=1)
train_df
```

```python
test_df = test_df.drop(['Name','Cabin','Ticket'], axis=1)
test_df
```

# Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier #Imports the Decision Tree Algorithm from sklearn
```

## Populating Independent and Dependent Variables

```python
x_train = train_df.drop("Survived", axis = 1) #Calls all features used for the decision tree except Survived as independent variables
y_train = train_df["Survived"] #the target variable
x_test = test_df.drop("PassengerId", axis = 1).copy()
```

```python
dt_model = DecisionTreeClassifier() #Establishes the decision tree algorithm
dt_model = dt_model.fit(x_train,y_train) #builds decision tree for the training set
y_pred = dt_model.predict(x_test) #predicts Survival for test set
```

## Graph Decision Tree

```python
from sklearn import tree
from six import StringIO
from IPython.display import Image
import pydotplus
```

```python
import graphviz
dot_data = tree.export_graphviz(dt_model, out_file=None, feature_names = ['Pclass','Gender','Age','SibSp','Parch','Fare','Embarked'], filled = True, rounded = True, special_characters = True) 
graph = graphviz.Source(dot_data)
graph
```

## Decision Tree Five-Fold Validation

```python
from sklearn.model_selection import cross_val_score #importing k-fold validation variable
```

```python
dt_fivefold = DecisionTreeClassifier()
dt_fivefold_score = cross_val_score(dt_fivefold, x_train, y_train, scoring = 'accuracy')

print("Decision Tree Scores: ", dt_fivefold_score)
print("Decision Tree Average: ",round(dt_fivefold_score.mean(),4))
```

# Random Forest Classifier Five-Fold Validation

```python
from sklearn.ensemble import RandomForestClassifier #imports Random Forest Algorithm
```

```python
rf_fivefold = RandomForestClassifier() #builds the random forest algorithm; from sklearn website, default number of trees = 100
rf_fivefold_score = cross_val_score(rf_fivefold, x_train, y_train, scoring = 'accuracy')

print("Random Forest Scores: ",rf_fivefold_score)
print("Random Forest Average: ", round(rf_fivefold_score.mean(),4))
```
